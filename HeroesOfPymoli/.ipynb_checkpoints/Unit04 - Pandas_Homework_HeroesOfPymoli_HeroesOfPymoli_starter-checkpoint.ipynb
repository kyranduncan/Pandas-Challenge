## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 100]

# Create labels for these bins
age_demographics_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]
```

```python
# Slice the data and place it into bins
pd.cut(purchase_data['Age'], bins, labels=age_demographics_labels)
```

```python
#put age ranges in dataframe to run % and count calculations

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels= age_demographics_labels, include_lowest=True)
purchase_data.head()
```

```python
# Create a GroupBy object based upon "Age Range"
age_group = purchase_data.groupby("Age Range")

# Find how many rows fall into each bin-count by age
count_by_age= age_group["SN"].nunique()
count_by_age
```

```python
#percent by age
percent_by_age= (count_by_age/unique_players)
percent_by_age.map("{:,.2%}".format)
```

```python
#DataFrame for count & percent by age
age_demographics = pd.DataFrame({"Total Count": count_by_age, "Percentage of Players": percent_by_age,})

# Format the data frame with no index name in the corner
age_demographics.index.name = None

# Format percentage with two decimal places 
age_demographics.style.format({"Percentage of Players":"{:,.2%}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Individual purchase data
players_purchase_count_df = purchase_data.groupby("SN").count()["Price"].rename("Purchase Count")
players_average_price_df = purchase_data.groupby("SN").mean()["Price"].rename("Average Purchase Price")
players_total_df = purchase_data.groupby("SN").sum()["Price"].rename("Total Purchase Value")

individual_purchase_data_df = pd.DataFrame({"Purchase Count":players_purchase_count_df,
                                   "Average Purchase Price": players_average_price_df,
                                   "Total Purchase Value": players_total_df})
individual_purchase_data_df.head(5)
```

```python
#top 5 spenders sort
top_five_spenders_df = individual_purchase_data_df.sort_values("Total Purchase Value", ascending=False)
top_five_spenders_df['Average Purchase Price']= top_five_spenders_df['Average Purchase Price'].map("${:,.2f}".format)
top_five_spenders_df['Total Purchase Value']= top_five_spenders_df['Total Purchase Value'].map("${:,.2f}".format)
top_five_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item purchase data
item_data=purchase_data.groupby(["Item ID", "Item Name"])
item_purchase_count_df = item_data['Item ID'].count()
item_price_df = item_data['Price'].unique().str[0]
item_purchase_value_df = item_data['Price'].sum()

item_analysis_df = pd.DataFrame({"Purchase Count":item_purchase_count_df,
                                   "Item Price": item_price_df,
                                   "Total Purchase Value": item_purchase_value_df})
                                
item_analysis_df.head()
```

```python
#Sort Item purchase data frame
item_analysis_df = item_analysis_df.sort_values("Purchase Count", ascending=False)
item_analysis_df['Item Price']= item_analysis_df['Item Price'].map("${:,.2f}".format)
item_analysis_df['Total Purchase Value']= item_analysis_df['Total Purchase Value'].map("${:,.2f}".format)
item_analysis_df.head()
```
